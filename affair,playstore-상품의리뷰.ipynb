- 두 데이터셋의 컬럼 구조를 통일시키고 합치고 싶음 

```python
import pandas as pd

# affairs와 playstore 데이터프레임 불러왔다고 가정
# 예: df_affairs, df_playstore

# affairs 컬럼: name, location, rating, date, review, helpful
# playstore 컬럼: name, rating, date, review, thumbsUp, replyContent, version, country

# 공통 컬럼 기준으로 맞추고, 나머지는 None 또는 NaN 추가
df_affairs['thumbsUp'] = None
df_affairs['replyContent'] = None
df_affairs['version'] = None
df_affairs['country'] = df_affairs['location'] if 'location' in df_affairs.columns else None
df_affairs.drop(columns=['location', 'helpful'], inplace=True, errors='ignore')

# 컬럼 순서 맞추기
target_columns = ['name', 'rating', 'date', 'review', 'thumbsUp', 'replyContent', 'version', 'country']
df_affairs = df_affairs[target_columns]
df_playstore = df_playstore[target_columns]

# 두 데이터프레임 병합
df_combined = pd.concat([df_affairs, df_playstore], ignore_index=True)
```

- 사용할 것만 잘 남겼고 이제 불용어 제거

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def clean_review_kor_en(text):
    # 소문자 변환
    text = text.lower()
    # 알파벳, 숫자, 공백 제외한 모든 문자 제거 (한글, 이모지, 특수문자 제거)
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    # 여러 공백 하나로
    text = re.sub(r'\s+', ' ', text).strip()
    # 불용어 제거
    words = [word for word in text.split() if word not in ENGLISH_STOP_WORDS]
    return ' '.join(words)
```

```python
df_use['cleaned_review'] = df_use['review'].astype(str).apply(clean_review_kor_en)
```

```python
df_use.drop(columns=['review'],inplace=True)
```

리뷰에 상품명이 등장하는지 확인해서 연결

```python
def find_mentioned_products(review, product_list):
    found = [product for product in product_list if product in review.lower()]
    return ', '.join(found) if found else None
```

```python
df_use['mentioned_products'] = df_use['review'].apply(lambda x: find_mentioned_products(str(x), product_names))
```

연결된 상품 리뷰만 추출 (필요 시)

```python
df_product_related = df_use[df_use['mentioned_products'].notnull()]
```
