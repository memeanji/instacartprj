```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec, Doc2Vec
from gensim.models.doc2vec import TaggedDocument
from nltk.corpus import stopwords

# ✅ stopwords만 사용 (nltk 다운로드 됐다고 가정)
stop_words = set(stopwords.words('english'))

# ✅ 오류 없는 안전 전처리 함수
def safe_preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()
    tokens = [word for word in words if word not in stop_words and len(word) > 2]
    return ' '.join(tokens)

# ✅ 전처리 적용
df['cleaned_review'] = df['review'].astype(str).apply(safe_preprocess)

# ✅ CountVectorizer
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(df['cleaned_review'])

# ✅ TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(df['cleaned_review'])

# ✅ Word2Vec
tokenized_reviews = df['cleaned_review'].apply(lambda x: x.split())
w2v_model = Word2Vec(sentences=tokenized_reviews, vector_size=100, window=5, min_count=1, workers=4)

def get_w2v_vector(tokens):
    valid_tokens = [token for token in tokens if token in w2v_model.wv]
    if valid_tokens:
        return sum(w2v_model.wv[token] for token in valid_tokens) / len(valid_tokens)
    else:
        return [0]*100

df['w2v_vector'] = tokenized_reviews.apply(get_w2v_vector)

# ✅ Doc2Vec
tagged_docs = [TaggedDocument(words=words, tags=[i]) for i, words in enumerate(tokenized_reviews)]
d2v_model = Doc2Vec(vector_size=100, window=5, min_count=1, workers=4, epochs=20)
d2v_model.build_vocab(tagged_docs)
d2v_model.train(tagged_docs, total_examples=d2v_model.corpus_count, epochs=d2v_model.epochs)
df['doc2vec_vector'] = [d2v_model.infer_vector(doc.words) for doc in tagged_docs]
```

1. 엘보우 기법 (Elbow Method) 🔽
- 군집 수 k에 따른 군집 내 제곱합(WCSS) 변화를 시각화

- WCSS가 급격히 감소하다 완만해지는 엘보우 지점을 최적 k로 선택

```python
# Word2Vec 훈련
from gensim.models import Word2Vec
import numpy as np

# 1. 'tokens' 컬럼 생성
df['tokens'] = df['cleaned_review'].apply(lambda x: x.split())

# 2. Word2Vec 훈련
w2v_model = Word2Vec(sentences=df['tokens'], vector_size=20, window=5, min_count=2, workers=4, sg=1, seed=42)

# 3. 평균 벡터 계산 함수
def get_avg_vector(tokens):
    vectors = [w2v_model.wv[token] for token in tokens if token in w2v_model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(20)

# 4. 'vector' 컬럼 생성
df['vector'] = df['tokens'].apply(get_avg_vector)

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

wcss = []
X = np.vstack(df['vector'].values)

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init='auto')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('클러스터 수 k')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid(True)
plt.show()
```

2. 실루엣 점수 (Silhouette Score)

```python
from sklearn.metrics import silhouette_score

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    print(f"k={k}, 실루엣 점수: {score:.4f}")

# 1에 가까울수록 좋은 군집화 
```

```python
#Word2Vec + KMeans 클러스터링 (20차원)
import pandas as pd
import gensim
from gensim.models import Word2Vec
from sklearn.cluster import KMeans
import numpy as np

# 1. 리뷰 텍스트 토큰화 (전처리 후라고 가정)
df['tokens'] = df['review'].astype(str).apply(lambda x: x.split())

# 2. Word2Vec 모델 훈련 (20차원)
w2v_model = Word2Vec(sentences=df['tokens'], vector_size=20, window=5, min_count=2, workers=4, sg=1, seed=42)

# 3. 리뷰 문장 벡터 계산 (토큰들의 평균 벡터)
def get_avg_vector(tokens):
    vectors = [w2v_model.wv[token] for token in tokens if token in w2v_model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(20)

df['vector'] = df['tokens'].apply(get_avg_vector)

# 4. KMeans 클러스터링 (엘보우가 애매해서 실루엣으로 결정했음)
k = 6
X = np.vstack(df['vector'].values)
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
df['cluster'] = kmeans.fit_predict(X)

# 5. 결과 확인
df[['review', 'cluster']].head()
```

- 이 결과를 시각화
- PCA + Plotly로 클러스터 시각화 코드

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import numpy as np
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# 1. 임베딩 벡터 준비
X = np.vstack(df['vector'].values)

# 2. PCA (2차원 축소)
pca = PCA(n_components=2, random_state=42)
pca_result = pca.fit_transform(X)

# 3. subplot 준비
k_list = [2, 3, 6, 8]
fig = make_subplots(rows=2, cols=2, subplot_titles=[f"k={k}" for k in k_list])

# 4. 각 K에 대해 클러스터링 및 시각화
for idx, k in enumerate(k_list):
    row = idx // 2 + 1
    col = idx % 2 + 1

    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    clusters = kmeans.fit_predict(X)
    sil_score = silhouette_score(X, clusters)

    for cluster_id in range(k):
        cluster_mask = clusters == cluster_id
        fig.add_trace(
            go.Scatter(
                x=pca_result[cluster_mask, 0],
                y=pca_result[cluster_mask, 1],
                mode='markers',
                marker=dict(size=4, opacity=0.6),
                name=f"k={k}, cluster {cluster_id}",
                showlegend=False
            ),
            row=row,
            col=col
        )
    
    # 제목에 실루엣 점수 추가
    fig.layout.annotations[idx].text = f"k={k} (실루엣: {sil_score:.3f})"

# 5. 전체 레이아웃 설정
fig.update_layout(
    title_text="KMeans 클러스터링 결과 (PCA 시각화)",
    height=800,
    width=1000,
    showlegend=False
)

fig.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from collections import defaultdict
import pandas as pd

# 1. 클러스터 수 설정 및 클러스터링 수행
k = 3
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
df['cluster'] = kmeans.fit_predict(np.vstack(df['vector'].values))

# 2. 토큰 -> 문자열로 다시 변환
df['joined_tokens'] = df['tokens'].apply(lambda x: ' '.join(x))

# 3. n-gram 추출 함수 정의
def get_top_ngrams(texts, ngram_range=(1,1), top_n=10):
    vec = CountVectorizer(ngram_range=ngram_range)
    X = vec.fit_transform(texts)
    sum_words = X.sum(axis=0)
    word_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    sorted_words = sorted(word_freq, key=lambda x: x[1], reverse=True)
    return sorted_words[:top_n]

# 4. 클러스터별로 n-gram 분석
cluster_ngrams = defaultdict(dict)

for i in range(k):
    cluster_texts = df[df['cluster'] == i]['joined_tokens']
    cluster_ngrams[i]['unigram']  = get_top_ngrams(cluster_texts, ngram_range=(1,1), top_n=10)
    cluster_ngrams[i]['bigram']   = get_top_ngrams(cluster_texts, ngram_range=(2,2), top_n=10)
    cluster_ngrams[i]['trigram']  = get_top_ngrams(cluster_texts, ngram_range=(3,3), top_n=10)

# 5. 결과 출력
for cluster_id, ngrams in cluster_ngrams.items():
    print(f"\n 📌클러스터 {cluster_id} 대표 n-grams:")
    for n in ['unigram', 'bigram', 'trigram']:
        print(f"  - {n}: {[w for w, _ in ngrams[n]]}")
```

```python
# SyntaxError: source code string cannot contain null bytes오류 났어서 정리하고 
import ast
import numpy as np

# 문자열이면 안전하게 파싱 + null byte 제거
def safe_parse(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x.replace('\x00', ''))
        except:
            return None  # 에러나는 경우는 None 처리
    return x

df['doc2vec_vector'] = df['doc2vec_vector'].apply(safe_parse)

# None 제거 후 배열 쌓기
df = df[df['doc2vec_vector'].notnull()]
doc2vec_vectors = np.vstack(df['doc2vec_vector'].values)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Doc2Vec 벡터 불러오기
doc2vec_vectors = np.vstack(
    df['doc2vec_vector'].apply(lambda x: ast.literal_eval(x.replace('\x00', '')) if isinstance(x, str) else x)
)

# t-SNE 차원 축소
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
tsne_embedding = tsne.fit_transform(doc2vec_vectors)

# KMeans 클러스터링
k = 3
kmeans = KMeans(n_clusters=k, random_state=42).fit(tsne_embedding)
df['tsne_cluster'] = kmeans.labels_

# 실루엣 점수
sil_score = silhouette_score(tsne_embedding, kmeans.labels_)

# 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(x=tsne_embedding[:, 0], y=tsne_embedding[:, 1],
                hue=df['tsne_cluster'], palette='Set2')
plt.title(f"t-SNE Clustering (Silhouette: {sil_score:.3f})")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.tight_layout()
plt.show()
```

- 위에서보다 pca 하고 kmeans했을때 실루엣 계수가 0.144였고 
- t-sne 해서 kmeans 했을때 실루엣 계수가 0.370으로 잘 보임

```python
# 실루엣 플롯(Silhouette Plot)을 클러스터 수(K)별로 비교하는 방법
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans

# 벡터 준비 (예: Doc2Vec, TF-IDF 등)
X = np.vstack(df['doc2vec_vector'].values)

# 클러스터 수 후보
range_n_clusters = [2, 3, 4, 5]

# 서브플롯 설정
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.flatten()

for idx, n_clusters in enumerate(range_n_clusters):
    # KMeans 학습
    clusterer = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
    cluster_labels = clusterer.fit_predict(X)

    # 실루엣 계수 계산
    silhouette_avg = silhouette_score(X, cluster_labels)
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    # 시각화
    ax = axs[idx]
    y_lower = 10
    for i in range(n_clusters):
        # 해당 클러스터의 실루엣 계수 추출
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax.fill_betweenx(np.arange(y_lower, y_upper),
                         0, ith_cluster_silhouette_values,
                         facecolor=color, edgecolor=color, alpha=0.7)

        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10  # 다음 클러스터 간 간격

    ax.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax.set_title(f"Number of Cluster : {n_clusters}\nSilhouette Score : {silhouette_avg:.3f}")
    ax.set_xlabel("The silhouette coefficient values")
    ax.set_ylabel("Cluster label")
    ax.set_yticks([])  # y축 눈금 제거
    ax.set_xlim([-0.1, 1])

plt.tight_layout()
plt.show()
```

-> 뭐가 다를 줄 알고 해본건데 다 같음 

- precision (정밀도): 예측한 것 중 실제로 맞은 비율
     - → "긍정이라고 한 것 중 진짜 긍정인 비율"
- recall (재현율): 실제 정답 중에서 예측이 맞은 비율
    - → "진짜 긍정을 놓치지 않고 잘 잡았나?"
- f1-score: precision과 recall의 조화 평균
    - → 두 지표가 균형잡히게 좋을수록 f1-score도 좋음
- support: 각 클래스의 실제 데이터 개수

| 클래스 | 의미 (예측 감정)   | precision | recall | f1-score | support |
| --- | ------------ | --------- | ------ | -------- | ------- |
| `0` | 긍정 리뷰 | 0.98      | 0.98   | 0.98     | 310개    |
| `2` | 부정 리뷰 | 0.95      | 0.94   | 0.94     | 110개    |


- 클래스 0 (긍정 리뷰)
    - precision 0.98 → 긍정이라고 한 것 중 98%가 실제 긍정!
    - recall 0.98 → 실제 긍정 중 98%를 잘 맞춤!

- 클래스 2 (부정 리뷰)
    - precision과 recall 모두 0.94~0.95 수준 → 아주 뛰어남
    - 다만 긍정 리뷰보다 약간 낮음 → 데이터 수가 적어서 그럴 수 있어요

| 지표               | 값    | 의미                             |
| ---------------- | ---- | ------------------------------ |
| **accuracy**     | 0.97 | 전체 리뷰 중 97% 정확히 맞춤             |
| **macro avg**    | 0.96 | 클래스 0, 2의 평균 성능 (클래스 수 균등하게 봄) |
| **weighted avg** | 0.97 | 각 클래스의 비율을 고려한 평균 성능           |
